```python
import pandas as pd
import numpy as np

pd.set_option('display.float_format', lambda x: '%.5f' % x)
np.set_printoptions(suppress=True,
   formatter={'float_kind':'{:f}'.format}) # Format array to remove e as best possible
import time   
import matplotlib.pyplot as plt
from IPython.display import Image
%matplotlib inline


from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.model_selection import train_test_split
from sklearn import tree

from sklearn.tree import export_graphviz 
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV #Searches through every combination and returns the best parameters to train the model
```

## 1. Loading & Cleansing of Data

```python
df = pd.read_csv("Math.csv")
```

```python
df['school'] = df['school'].map({'GP': 1,'MS': 0}).astype(int)

df['sex'] = df['sex'].map({'M': 1,'F': 0}).astype(int)

df.loc[(df['age'] >= 15 ) & (df['age'] <=18), 'age'] = 0
df.loc[(df['age'] > 18) & (df['age'] <=22), 'age'] = 1
df['age'] = df['age'].astype(int)

df['address'] = df['address'].map({'U': 1,'R': 0}).astype(int)

df['famsize'] = df['famsize'].map({'LE3': 0,'GT3': 1}).astype(int)

df['Pstatus'] = df['Pstatus'].map({'T': 1,'A': 0}).astype(int)

df['Mjob'] = df['Mjob'].map({'teacher': 0,'health': 1,'services': 2,'at_home': 3,'other': 4}).astype(int)

df['Fjob'] = df['Fjob'].map({'teacher': 0,'health': 1,'services': 2,'at_home': 3,'other': 4}).astype(int)

reason_mapping = {"home":1,"reputation":2,"course":3,"other":4}
df['reason'] = df['reason'].map(reason_mapping).astype(int)

guardian_mapping = {"mother":1,"father":2,"other":3}
df['guardian'] = df['guardian'].map(guardian_mapping).astype(int)

df['failures'] = df['failures'].apply(lambda x: x if 1<=x<3 else 4).astype(int)

schoolsup_mapping = {"yes":1,"no":0}
df['schoolsup'] = df['schoolsup'].map(schoolsup_mapping).astype(int)

famsup_mapping = {"yes":1,"no":0}
df['famsup'] = df['famsup'].map(famsup_mapping).astype(int)

paid_mapping = {"yes":1,"no":0}
df['paid'] = df['paid'].map(paid_mapping).astype(int)

activities_mapping = {"yes":1,"no":0}
df['activities'] = df['activities'].map(activities_mapping).astype(int)

nursery_mapping = {"yes":1,"no":0}
df['nursery'] = df['nursery'].map(nursery_mapping).astype(int)

#Encode 'higher' column #21
df['higher'] = df['higher'].map({'yes':1, 'no': 0}).astype(int)
#Encode 'internet' column #22
df['internet'] = df['internet'].map({'yes':1, 'no': 0}).astype(int)
#Encode 'higher' column #23
df['romantic'] = df['romantic'].map({'yes':1, 'no': 0}).astype(int)
```

```python
# Transform data
df.loc[df['G3'] <= 9, 'G3'] = 0
df.loc[(df['G3'] >= 10) & (df['G3']<=11),'G3'] = 1
df.loc[(df['G3'] >= 12) & (df['G3']<=13),'G3'] = 2
df.loc[(df['G3'] >= 14) & (df['G3']<=15),'G3'] = 3
df.loc[df['G3'] > 15, 'G3'] = 4
```

```python
# Set the "G3" as target/model output and the rest features as model inputs
dfX = df.iloc[:,0:32]
dfy = df.iloc[:,-1]
X = dfX.values
y = dfy.values
print("X shape is: {} and type is {}".format(X.shape,type(X)))
print("y shape is: {} and type is {}".format(y.shape,type(y)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

## 2. Inital Decision Tree Regressor (All Columns Are Used)

```python
# Create Decision Tree Model
tree_reg = tree.DecisionTreeRegressor(max_depth=2, random_state=2)
tree_reg.fit(X_train, y_train)
```

```python
# Export our trained model as a .dot file
export_graphviz(
        tree_reg,
        out_file=("regression_tree.dot"),
        feature_names = list(df.iloc[:,0:32].columns),
                rounded=True,
        filled=True
    )
```

```python
#training mse
train_mse = mean_squared_error(tree_reg.predict(X_train), y_train)
print('The training mean squared error is: ', train_mse)
#testing mse
test_mse = mean_squared_error(tree_reg.predict(X_test), y_test)
print('The testing mean squared error is: ',test_mse)
```

```python
#the score function for regressor is calculating R Squared Value of the regression model
# R Squared value rangs from 0 to 1 and the higher the better
train_R2 = tree_reg.score(X_train, y_train)
print('training R^2 value is: ', train_R2)

test_R2 = tree_reg.score(X_test, y_test)
print('testing R^2 value is: ', test_R2)
```

### K-Fold Cross Validation

```python
# Evaluate the Model using K-folds Cross validation
results  = cross_validate(tree_reg, X, y.ravel(), scoring='neg_mean_squared_error', cv=5, return_train_score = True)
print('train_score: ', results['train_score'])
print('test_score: ', results['test_score'])
```

```python
print('cross val training mean_squared_error is:', sum(-results['train_score'])/len(results['train_score']))
print('cross val testing mean_squared_error is:', sum(-results['test_score'])/len(results['test_score']))
```

## 3.Grid Search

```python
start = time.time()

depth_range = range(0,10)

train_mse = []
test_mse = []
cross_val_train_mse =[]
cross_val_test_mse =[]

for d in depth_range:
    train_mse.append(mean_squared_error(tree_reg.predict(X_train), y_train))
    test_mse.append(mean_squared_error(tree_reg.predict(X_test), y_test))
    results  = cross_validate(tree_reg, X, y.ravel(), scoring='neg_mean_squared_error', cv=5, return_train_score = True)
    cross_val_train_mse.append(sum(-results['train_score'])/len(results['train_score']))
    cross_val_test_mse.append(sum(-results['test_score'])/len(results['test_score']))
    
end = time.time()
time_taken = end - start
print('Time: ',time_taken)
```

```python
# Plot the simple split measures (train_mse & test_mse) 
plt.plot(depth_range, train_mse, 'bo-', label ='training mse')
plt.plot(depth_range, test_mse, 'ro-', label = 'testing mse')
#plt.figure(figsize=(,))
plt.xlabel('max_depth', fontsize='x-large')
plt.ylabel('mean squared error (MSE)', fontsize='x-large')

plt.legend(loc='best', shadow=True, fontsize='x-large')
plt.show()
```

```python
# Create the final Decision Tree Regressor using the best hyperparameters
tree_reg = tree.DecisionTreeRegressor(criterion="mae",max_depth=4,min_samples_leaf=14,min_samples_split=2)
tree_reg.fit(X_train,y_train)
```

```python
# Task 5: Evaluate the Model by calculating training mse and testing mse

#trainning mse
train_mse = mean_squared_error(tree_reg.predict(X_train),y_train)
print("The trainning mean squared error is: ",train_mse)

#testing mse
test_mse = mean_squared_error(tree_reg.predict(X_test),y_test)
print("The testing mean squared error is: ", test_mse)

#testing mae
test_mse = mean_squared_error(tree_reg.predict(X_test),y_test)
print("The testing mean squared error is: ", test_mse)
```

```python
#the score function for regressor is calculating R Squared Value of the regression model
# R Squared value rangs from 0 to 1 and the higher the better
train_R2 = tree_reg.score(X_train, y_train)
print('training R^2 value is: ', train_R2)

test_R2 = tree_reg.score(X_test, y_test)
print('testing R^2 value is: ', test_R2)
```
